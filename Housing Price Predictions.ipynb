```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
import bs4 as bs
import urllib.request

from scipy import stats
from pathlib import Path
```

```python
source = urllib.request.urlopen('https://dqydj.com/historical-home-prices/').read()
soup = bs.BeautifulSoup(source,'lxml')
```

```python
# title of the page
print(soup.title)

#extract tale data for housing prices
table = soup.table
table = soup.find('table')
table_rows = table.find_all('tr')
for tr in table_rows:
    td = tr.find_all('td')
    row = [i.text for i in td]
    print(row)
```
